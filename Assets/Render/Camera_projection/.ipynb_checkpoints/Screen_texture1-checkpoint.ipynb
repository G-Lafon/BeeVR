```python
import numpy as np
import matplotlib.pyplot as plt
import imageio

imageio.plugins.freeimage.download()
```

```python
h=40*4
w=80*4*6
#h=w

Amax = np.pi/6
hReal = 320
distanceBeeScreen = 300 

x = 2*np.pi*(np.arange(0,w)-w/2)/w
#y = np.arctan2(hReal*((np.arange(0,h)-h/2)/(h)),distanceBeeScreen)
A0 =  np.arctan2(8,60)
A0 = np.pi
y = A0*((np.arange(0,h)-h/2)/(h))
```

```python
X = np.tile(x,(h,1))
Y = np.tile(y,(w,1)).T

```

```python
X[:,int(w/2):] = np.flipud(np.fliplr(X[:,int(w/2):]))
Y[:,int(w/2):] = np.flipud(Y[:,int(w/2):])
```

```python
texture = np.zeros((h,w,3)).astype("float32")
texture[:,:,0] = np.cos(X)*np.cos(Y)
texture[:,:,2] = np.sin(X)*np.cos(Y)
texture[:,:,1] = np.sin(Y)
```

```python
plt.imshow((texture+1)/2)
```

```python

imageio.imwrite(r".\cylinderMap7.exr",(texture+1)/2)
```
