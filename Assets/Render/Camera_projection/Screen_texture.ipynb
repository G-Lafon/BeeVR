```python
import numpy as np
import matplotlib.pyplot as plt
import imageio

imageio.plugins.freeimage.download()
```

```python
h=40*4
w=80*6*4


Amax = np.pi/6
hReal = 320
distanceBeeScreen = 300 

x = 2*np.pi*(np.arange(0,w)-w/2)/w
y = np.pi*((np.arange(0,h)-h/2)/(h))
```

```python
X = np.tile(x,(h,1))
Y = np.tile(y,(w,1)).T

```

```python
plt.imshow(Y)
```

```python
texture = np.zeros((h,w,3)).astype("float32")
texture[:,:,0] = np.cos(X)*np.cos(Y)
texture[:,:,2] = np.sin(X)*np.cos(Y)
texture[:,:,1] = np.sin(Y)
```

```python
plt.imshow((texture+1)/2)
```
